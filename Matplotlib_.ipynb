Matplotlib 플로팅

x 및 y 점 표시
이 plot()함수는 다이어그램에 점(마커)을 그리는 데 사용됩니다.

title()Pyplot을 사용하면 플롯의 제목을 설정하는 함수를 사용할 수 있습니다 .

x축과 y축에 대한 플롯 제목과 레이블을 추가합니다.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

제목 및 레이블에 대한 글꼴 속성 설정


제목과 레이블의 글꼴 속성을 설정하려면 , , fontdict의 매개변수 를 사용할 수 있습니다 .xlabel()ylabel()title()

제목과 레이블에 대한 글꼴 속성을 설정합니다.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

Matplotlib 서브플롯

여러 개의 플롯 표시

이 기능을 사용하면 subplot()하나의 그림에 여러 개의 플롯을 그릴 수 있습니다.

2개의 플롯을 그리세요:

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

subplot() 함수
이 subplot()함수는 그림의 레이아웃을 설명하는 세 개의 인수를 사용합니다.

레이아웃은 행과 열로 구성되며, 이는 첫 번째 와 두 번째 인수로 표현됩니다.

세 번째 인수는 현재 플롯의 인덱스를 나타냅니다.


plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.

plt.subplot(1, 2, 2)
#the figure has 1 row, 2 columns, and this plot is the second plot.
따라서 행 2개, 열 1개로 구성된 그림(두 개의 플롯이 나란히 표시되는 대신 서로 위에 겹쳐서 표시됨)이 필요한 경우 구문을 다음과 같이 작성할 수 있습니다.

2개의 플롯을 서로 위에 그리세요:

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

산점도 만들기

Pyplot을 사용하면 scatter()산점도를 그릴 수 있습니다.

이 scatter()함수는 각 관측값에 대해 하나의 점을 표시합니다. 길이가 같은 두 개의 배열이 필요한데, 하나는 x축 값을 위한 것이고 다른 하나는 y축 값을 위한 것입니다.

예

간단한 산점도:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

위의 예에서 관찰한 결과는 13대의 자동차가 지나간 결과입니다.

X축은 자동차의 연식을 나타냅니다.

Y축은 자동차가 지나갈 때의 속도를 보여줍니다.

관찰 결과 사이에 관계가 있나요?

자동차가 최신형일수록 더 빨리 달리는 것 같지만, 그것은 우연일 수도 있습니다. 결국 우리는 단 13대의 자동차만 등록했으니까요.

플롯 비교

위의 예에서는 속도와 나이 사이에 관계가 있는 것처럼 보이지만, 다른 날의 관측치도 함께 표시해 보면 어떨까요? 산점도가 다른 정보를 알려주는 걸까요?

컬러맵


Matplotlib 모듈에는 다양한 컬러맵이 있습니다.

컬러맵은 각 색상이 0에서 100까지의 값을 갖는 색상 목록과 같습니다.

컬러맵의 예는 다음과 같습니다.



이 컬러맵은 'viridis'라고 불리며, 보시다시피 보라색인 0부터 노란색인 100까지 범위가 있습니다.

ColorMap을 사용하는 방법
키워드 인수에 cmap컬러맵 값을 지정하여 컬러맵을 지정할 수 있습니다. 이 경우 컬러 맵 'viridis'은 Matplotlib에서 제공하는 기본 컬러맵 중 하나입니다.

또한 산점도의 각 지점에 대해 하나의 값, 즉 0~100까지의 값을 갖는 배열을 만들어야 합니다.

예

색상 배열을 만들고 산점도에서 색상 맵을 지정합니다.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.show()
```

다음 문장을 포함하면 도면에 색상 맵을 포함할 수 있습니다 plt.colorbar().

파이 차트 만들기

Pyplot을 사용하면 다음 pie()함수를 사용하여 원형 차트를 그릴 수 있습니다.

예

간단한 원형 차트:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

보시다시피 원형 차트는 배열의 각 값(이 경우 [35, 25, 25, 15])에 대해 하나의 조각(쐐기라고 함)을 그립니다.

기본적으로 첫 번째 쐐기의 플로팅은 x축에서 시작하여 시계 반대 방향으로 이동합니다 .

라벨

매개변수를 사용하여 원형 차트에 레이블을 추가합니다 labels.

매개 labels변수는 각 쐐기마다 하나의 레이블이 있는 배열이어야 합니다.

예

간단한 원형 차트:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

시작 각도
앞서 언급했듯이 기본 시작 각도는 x축에 있지만 startangle매개변수를 지정하여 시작 각도를 변경할 수 있습니다.

매개 startangle변수는 각도로 정의되며 기본 각도는 0입니다.

예

첫 번째 쐐기를 90도에서 시작하세요.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

터지다

웨지 중 하나를 눈에 띄게 하고 싶으신가요? explode매개변수를 사용하면 가능합니다.

매개 explode변수가 지정되면 None각 쐐기마다 하나의 값을 갖는 배열이어야 합니다.

각 값은 각 쐐기가 중심에서 얼마나 떨어져 표시되는지를 나타냅니다.

예

파이 중앙에서 "사과" 조각을 0.2만큼 떼어냅니다.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

그림자

shadows매개변수를 다음과 같이 설정하여 원형 차트에 그림자를 추가합니다 True.

예

그림자 추가:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```

그림 물감

매개변수를 사용하여 각 쐐기의 색상을 설정할 수 있습니다 colors.

매개 colors변수가 지정된 경우 각 쐐기마다 하나의 값을 갖는 배열이어야 합니다.

예

각 쐐기마다 새로운 색상을 지정하세요.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show()
```

16진수 색상 값 , 지원되는 140개 색상 이름 중 하나 또는 다음 단축키 중 하나를 사용할 수 있습니다 .

'r'- 빨간색
'g'- 초록색
'b'- 파란색
'c'- 청록색
'm'- 자홍색
'y'- 노란색
'k'- 검은색
'w'- 흰색

전설

각 쐐기에 대한 설명 목록을 추가하려면 다음 legend()함수를 사용하세요.

예

범례 추가:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show()
```

헤더가 있는 범례

범례에 헤더를 추가하려면 함수 title에 매개변수를 추가하세요 legend .

예

헤더가 있는 범례를 추가합니다.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show()
```
